```python
import math

def rotate_image(image, angle_degrees):
    """
    Rotate the image counterclockwise by the given angle using nearest neighbor interpolation.
    """
    # Write code here
    H = len(image)
    W = len(image[0])

    # Center of the image
    cy = (H-1)/2
    cx = (W-1)/2

    output = [[0 for _ in range(W)] for _ in range(H)]
    for i in range(H):
        for j in range(W):
            dy = i - cy 
            dx = j - cx
            src_y = cy + dy*math.cos(angle_degrees) + dx*math.sin(angle_degrees)
            src_x = cx + dy*math.sin(angle_degrees) + dx*math.cos(angle_degrees)
            src_y = int(round(src_y))
            src_x = int(round(src_x))
            if src_x <= W and src_y <= H:
                output[i][j] = image[src_x][src_y]
    return output


image = [[1,2,3],[4,5,6],[7,8,9]]
angle_degrees = 0
```

```python
output = rotate_image(image, angle_degrees)
output
```

```python
import math 
def bilinear_resize(image, new_h, new_w):
    """
    Resize a 2D grid using bilinear interpolation.
    """
    # Write code here
    height = len(image)
    width = len(image[0])
    output = [[0.0 for _ in range(height)] for _ in range(width)]
    for i in range(height):
        for j in range(width): 
            src_y = i * ((height-1)/(new_h-1))
            src_x = j * ((width-1)/(new_w-1))

            y_0 = math.floor(src_y)
            x_0 = math.floor(src_x)

            dy = src_y - y_0
            dx = src_x - x_0

            y_1 = min(y_0 + 1, height - 1)
            x_1 = min(x_0 + 1, width - 1)

            result = image[y_0][x_0]*(1-dy)(1-dx) + image[y_1][x_0]*dy*(1-dx) + image[y_0][x_1] * (1-dy)*dx + image[y_1][x_1]*dy*dx

            output[i][j] = result
    return output 
            
image = [[0, 10], [20, 30]] 
new_h = 3
new_w = 3
```

```python
import numpy as np
import math
def percentiles(x, q):
    """
    Compute percentiles using linear interpolation.
    """
    result = []
    x_sorted = sorted(x)
    n_data = len(x_sorted)
    n_pen = len(q) 
    for i in range(n_pen):
        L = q[i]/100 * (n_data+1)
        if L.is_integer():
            result.append(x_sorted[int(L)])
            print(x_sorted[int(L)])
        else:
            x_1 = int(math.floor(L))
            x_2 = x_1 + 1 
            percent = x_sorted[x_1-1] + 0.5 * (x_sorted[x_2-1] - x_sorted[x_1-1])
            result.append(percent)
    return np.array(result)
x=[1,2,3,4]
q=[15, 20, 25, 40, 50]
percentiles(x, q)
```

```python
import numpy as np

def adamw_step(w, m, v, grad, lr=0.001, beta1=0.9, beta2=0.999, weight_decay=0.01, eps=1e-8):
    """
    Perform one AdamW update step.
    """
    # Write code here
    m_t = beta1*m + (1-beta1)*grad
    v_t = beta2*v + (1-beta2)*grad**2
    w = w - lr*(weight_decay*w) - lr*(m/(np.sqrt(v)+eps))
    return m_t, v_t, w
w = [1.0, -2.0]
m = [0.0, 0.0]
v = [0.0, 0.0]
grad = [0.3, -0.7]
lr=0.01
weight_decay=0
adamw_step(w, m, v, grad, lr=0.01, weight_decay=0)
```

```python
import numpy as np

def nadam_step(w, m, v, grad, lr=0.002, beta1=0.9, beta2=0.999, eps=1e-8):
    """
    Perform one Nadam update step.
    """
    m_np = np.array(m)
    v_np = np.array(v)
    grad_np = np.array(grad)
    w_np = np.array(w)
    
    m_t =  beta1 * m_np + (1-beta1)*grad_np
    v_t = beta2* v_np + (1-beta2)*grad_np**2
    score = (beta1*m_t + (1-beta1)*grad_np)/(np.sqrt(v_t)+eps)
    w_score = w_np - lr*score
    return w_score

w=[1.0, -1.0]
m=[0.1, -0.1]
v=[0.01, 0.01]
grad=[0.2, -0.3]
lr=0.002

nadam_step(w, m, v, grad, lr=0.002)
```

```python
import numpy as np

def matrix_normalization(matrix, axis=None, norm_type='l2'):
    """
    Normalize a 2D matrix along specified axis using specified norm.
    """
    matrix_np = np.asarray(matrix, dtype=float)

    if norm_type == "l2":
        norm = np.sqrt(np.sum(matrix_np ** 2, axis=axis, keepdims=True))

    elif norm_type == "l1":
        norm = np.sum(np.abs(matrix_np), axis=axis, keepdims=True)

    elif norm_type == "max":
        norm = np.max(np.abs(matrix_np), axis=axis, keepdims=True)
    # tránh chia cho 0
    norm = np.where(norm == 0, 1, norm)

    return np.round(matrix_np / norm, 2)
matrix= [[2,8],[4,2]]
axis=1
norm_type='max'
```
